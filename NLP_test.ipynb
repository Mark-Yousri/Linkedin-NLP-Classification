The default pipeline for the clean method is the following:

fillna(s) Replace not assigned values with empty spaces.

lowercase(s) Lowercase all text.

remove_digits() Remove all blocks of digits.

remove_punctuation() Remove all string.punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~).

remove_diacritics() Remove all accents from strings.

remove_stopwords() Remove all stop words.

remove_whitespace() Remove all white space between words.

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

def Lemming(tokenized_text):
    text = [wordnet_lemmatizer.lemmatize(word) for word in tokenized_text]
    return text

df['job_requirements_filtered'] = df['job_requirements_filtered'].apply(Lemming)
df['company_information_filtered'] = df['company_information_filtered'].apply(Lemming)
df['job_description_filtered'] = df['job_description_filtered'].apply(Lemming)
```

```python
from nltk import ngrams

# Split text into tokens
df['job_requirements_filtered_tokens'] = df['job_requirements_filtered'].str.split()
df['company_information_filtered_tokens'] = df['company_information_filtered'].str.split()
df['job_description_filtered_tokens'] = df['job_description_filtered'].str.split()

# Generate n-grams (n=3 for trigrams)
df['job_requirements_filtered_trigrams'] = df['job_requirements_filtered_tokens'].apply(lambda x: list(ngrams(x, 3)))
df['company_information_filtered_trigrams'] = df['company_information_filtered_tokens'].apply(lambda x: list(ngrams(x, 3)))
df['job_description_filtered_trigrams'] = df['job_description_filtered_tokens'].apply(lambda x: list(ngrams(x, 3)))
```
