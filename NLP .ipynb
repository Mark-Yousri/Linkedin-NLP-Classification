The default pipeline for the clean method is the following:

fillna(s) Replace not assigned values with empty spaces.

lowercase(s) Lowercase all text.

remove_digits() Remove all blocks of digits.

remove_punctuation() Remove all string.punctuation (!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~).

remove_diacritics() Remove all accents from strings.

remove_stopwords() Remove all stop words.

remove_whitespace() Remove all white space between words.

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()

def Lemming(tokenized_text):
    text = [wordnet_lemmatizer.lemmatize(word) for word in tokenized_text]
    return text

df['job_requirements_filtered'] = df['job_requirements_filtered'].apply(Lemming)
df['company_information_filtered'] = df['company_information_filtered'].apply(Lemming)
df['job_description_filtered'] = df['job_description_filtered'].apply(Lemming)
```

```python
dfr = pd.DataFrame(index=df.index)
le = preprocessing.LabelEncoder()
for col in ['job_requirements_filtered_trigrams', 
            'company_information_filtered_trigrams', 
            'job_description_filtered_trigrams']:
    df[col] = df[col].apply(lambda x: ','.join(['_'.join(map(str, t)) for t in x]) if isinstance(x, list) else x)
    dfr[col] = le.fit_transform(df[col])
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['fake?','job_requirements_filtered_tf_idf','company_information_filtered_tf_idf','job_description_filtered_tf_idf'],axis=1)
y=df['fake?']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.35, random_state =  101)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state =83 ,max_iter=3000)
classifier.fit(X_train, y_train)
```

```python
y_pred_2 = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred_2)
print (cm)
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_test, y_pred_2)
print(cr)
```

```python
from sklearn.svm import SVC
svc_model = SVC(C=150, kernel='poly', degree= 4)
svc_model.fit(X_train, y_train)
print(svc_model.score(X_train, y_train))
print(svc_model.score(X_test, y_test))
y_pred = svc_model.predict(X_test)

# Generate a classification report for the true and predicted labels
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
dtclf = DecisionTreeClassifier(max_depth=3, criterion='gini',
                               random_state=100)
clf = AdaBoostClassifier(base_estimator=dtclf,
                            n_estimators=60,
                            learning_rate=0.8,
                            algorithm='SAMME.R',
                            random_state=100)

clf.fit(X_train, y_train)
print('Model accuracy score: {0:0.4f}'. format(clf.score(X_train,y_train)*100))
print('Model accuracy score: {0:0.4f}'. format(clf.score(X_test,y_test)*100))
```

```python
from sklearn.neural_network import MLPClassifier
MLP = MLPClassifier(hidden_layer_sizes=(20,5,30), activation='relu')
MLP.fit(X_train,y_train)
```

```python
print('Model accuracy score: {0:0.4f}'. format(MLP.score(X_train,y_train)*100))
print('Model accuracy score: {0:0.4f}'. format(MLP.score(X_test,y_test)*100))
```

```python
y_pred = MLP.predict(X_test)
MLP_report = classification_report(y_test, y_pred)
print(MLP_report)
```

```python
from xgboost import XGBClassifier
XGB = XGBClassifier(max_depth=5, subsample=1, n_estimators=50, min_child_weight=1, random_state=5)
XGB.fit(X_train, y_train)
print('Model accuracy score: {0:0.4f}'. format(XGB.score(X_train,y_train)*100))
print('Model accuracy score: {0:0.4f}'. format(XGB.score(X_test,y_test)*100))
```

```python
y_pred = XGB.predict(X_test)
XGB_report = classification_report(y_test, y_pred)
print(XGB_report)
```
